```python
%pip install pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
try:
    dataAdd = pd.read_csv('time_addVertex.dat', sep=' ', header=None, names=['n', 'time'])
    dataDel = pd.read_csv('time_removeVertex.dat', sep=' ', header=None, names=['n', 'time'])
    addX_values = dataAdd['n'].tolist()
    addY_values = dataAdd['time'].tolist()
    delX_values = dataDel['n'].tolist()
    delY_values = dataDel['time'].tolist()

    plt.figure(figsize=(20, 12))
    plt.plot(addX_values, addY_values, label='addVertex', marker='o')
    plt.title('Time Complexity of addVertex')
    plt.xlabel('Number of vertices')
    plt.ylabel('Time (ns)')
    plt.ylim(0, 100)
    plt.legend()
    plt.grid(True)
    plt.savefig('addVertex.png')

    plt.figure(figsize=(20, 12))
    plt.plot(delX_values, delY_values, label='delVertex', marker='o')
    plt.title('Time Complexity of delVertex')
    plt.xlabel('Number of vertices')
    plt.ylabel('Time (ns)')
    plt.ylim(1000, 2000)  # <--- DODANA LINIA
    plt.legend()
    plt.grid(True)
    plt.savefig('delVertex.png')

except Exception as e:
    print(f"Error {e}")

```
